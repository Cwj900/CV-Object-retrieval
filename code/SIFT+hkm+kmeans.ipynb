```python
import os
data_path=r"oxbuild_images"

def get_image_paths(data_dir):
    image_paths=[] #图片地址

    for image_dir in os.listdir(data_dir):
        image_path=os.path.join(data_dir, image_dir) 
        image_paths.append(image_path)
 
    return image_paths

# 加载训练集和测试集
image_paths= get_image_paths(data_path)

print("样本数量：", len(image_paths))
print(image_paths)
```

```python
import cv2
import numpy as np
import pickle

sift = cv2.SIFT_create()

# 提取图像特征
def extract_sift_features(image_path):
    image = cv2.imread(image_path)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    keypoints, descriptors = sift.detectAndCompute(gray_image, None)
    if descriptors is not None:
        return descriptors
    return None

# 提取第一张图像的特征
first_image_features = extract_sift_features(image_paths[0])

# 打印特征
print(first_image_features)

# 存储数据
with open('first_image_features.pkl', 'wb') as f:
    pickle.dump(first_image_features, f)
```

```python
import cv2
import numpy as np

sift = cv2.SIFT_create()

# 提取图像特征
def extract_sift_features(image_paths):
    all_features = []  # 这将是一个三维列表
    for path in image_paths:
        image = cv2.imread(path)
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        keypoints, descriptors = sift.detectAndCompute(gray_image, None)
        if descriptors is not None:
            all_features.append(descriptors.tolist())  # 将numpy数组转换为列表
        else:
            all_features.append([])  # 如果没有描述符，添加一个空列表
    return all_features


# 提取图像特征
features_per_image = extract_sift_features(image_paths)

# 将特征保存为.npy文件
np.save('image_features_list.npy', np.array(features_per_image, dtype=object))
```

```python
import cv2
import numpy as np
from sklearn.cluster import MiniBatchKMeans

# 初始化SIFT检测器
sift = cv2.SIFT_create()

# 提取图像特征
def extract_sift_features(image_paths):
    features = []
    for path in image_paths:
        image = cv2.imread(path)
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        keypoints, descriptors = sift.detectAndCompute(gray_image, None)
        if descriptors is not None:
            features.extend(descriptors)
    return features

# 聚类特征
def clustering(features, num_clusters):
    features = np.array(features)
    kmeans = MiniBatchKMeans(n_clusters=num_clusters, random_state=0, batch_size=100, max_iter=10)
    kmeans.fit(features)
    visual_words = kmeans.cluster_centers_
    return visual_words

# 提取图像特征
features = extract_sift_features(image_paths)

# 假设我们想要的视觉词汇大小为 500
num_visual_words = 500

# 聚类特征以创建视觉词汇
visual_words = clustering(features, num_visual_words)

print("视觉词汇构建完成，词汇大小为:", len(visual_words))
```

```python
# 保存每张图片的SIFT特征和视觉词汇
np.save('visual_words.npy', visual_words)
np.save('image_features_list.npy', features)
```

```python
import faiss
import numpy as np

def l1_distance(x, y):
    return np.abs(x - y).sum()

def hkm_clustering_faiss(features, num_clusters):
    feature_length = len(features[0])
    index = faiss.IndexIVFFlat(faiss.IndexFlatL2(feature_length), feature_length, faiss.METRIC_L1)  # 使用L1距离

    # 将特征向量添加到索引中
    index.train(np.array(features).astype('float32'))
    index.add(np.array(features).astype('float32'))

    # 初始化聚类中心为随机选择的特征向量
    cluster_centers = np.array(features[np.random.choice(len(features), num_clusters, replace=False)])

    while len(cluster_centers) > num_clusters:
        print(f"Current number of clusters: {len(cluster_centers)}")
        num_clusters = len(cluster_centers)
        new_cluster_centers = []
        new_cluster_assignments = []

        # 分配步骤
        D, I = index.search(np.array(cluster_centers).astype('float32'), 1)  # 找到每个聚类中心最近的特征向量
        for j, idx in enumerate(I.flatten()):
            cluster = features[np.where(I.flatten() == idx)]
            new_cluster_centers.append(np.mean(cluster, axis=0))
            new_cluster_assignments.extend([j] * len(cluster))

        # 更新步骤
        cluster_centers = np.array(new_cluster_centers)
        index.reset()
        index.train(np.array(cluster_centers).astype('float32'))
        index.add(np.array(cluster_centers).astype('float32'))

    return cluster_centers

# 加载特征
features = np.load('image_features_list.npy').astype('float32')

# 视觉词汇大小
num_visual_words = 1000

# 使用HKM进行聚类
visual_words_hkm = hkm_clustering_faiss(features, num_visual_words)
print("HKM视觉词汇构建完成，词汇大小为:", len(visual_words_hkm))
```
